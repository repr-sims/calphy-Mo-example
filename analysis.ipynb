```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp_sol, fe_sol = np.loadtxt('ts-bcc-solid-2500-0/temperature_sweep.dat', unpack=True, usecols=(0,1))
temp_lqd, fe_lqd = np.loadtxt('ts-bcc-liquid-2500-0/temperature_sweep.dat', unpack=True, usecols=(0,1))
```

```python
arg = np.argsort(np.abs(fe_sol - fe_lqd))[0]
print(temp_sol[arg])
```

```python
plt.plot(temp_sol, fe_sol, label='solid')
plt.plot(temp_lqd, fe_lqd, label='liquid')
plt.axvline(temp_sol[arg], color='k', linestyle='--')
plt.xlabel('Temperature (K)')
plt.ylabel('Free energy (eV/atom)')
```
